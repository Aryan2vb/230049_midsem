# Task 2.1 — Dataset Selection and Setup

```python
import numpy as np
import random
import pandas as pd
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

np.random.seed(42)
random.seed(42)
print('Imports and seeds initialized')
```

**Seed cell:** All random seeds are fixed at the top of this notebook for reproducibility.

```python
from sklearn.datasets import make_blobs

X, y_true = make_blobs(n_samples=300, centers=4, n_features=2,
                       cluster_std=0.8, random_state=42)
df = pd.DataFrame(X, columns=['x1', 'x2'])
df['label'] = y_true
df.to_csv('data/dataset.csv', index=False)

print(f"Shape: {X.shape}")
print(f"Classes: {np.unique(y_true)}")
print(df.head())
```

**Dataset justification (3–5 sentences):**

The dataset is a synthetic 2D dataset generated with `sklearn.datasets.make_blobs` — 300 samples drawn from 4 isotropic Gaussian blobs with standard deviation 0.8.

This is a reasonable testbed for DP-means because `make_blobs` generates spherical Gaussian clusters, which directly satisfies DP-means' core assumption that clusters are isotropic (Section 4, Eq. 9). The 2D structure allows clear visual verification of whether the algorithm correctly recovers k=4 and assigns points to the right clusters, making it easy to inspect both success and failure cases.

A key limitation compared to the paper's original datasets is that `make_blobs` is idealized — it contains no outliers, noise points, or overlapping structure, and the clusters are artificially balanced in size. The paper uses real-world datasets including image patches and speaker data (Section 5), which have irregular geometry and high dimensionality that stress the algorithm more than this toy setting does.

**Preprocessing:** No preprocessing required.


```python
print("Dataset Statistics:")
print(df.describe())
```

Basic statistics confirm the dataset has 4 balanced classes across two continuous features — meeting the exam requirement of ≥100 samples and ≥2 features. This empirical confirmation of the dataset properties is a fundamental prerequisite before applying the distance computation step (Algorithm 1, line 4) under the assumption of equal variance (Eq. 9).

```python
colors = plt.cm.tab10(np.linspace(0, 0.4, 4))
fig, ax = plt.subplots(figsize=(7, 5))
for j in range(4):
    mask = y_true == j
    ax.scatter(X[mask, 0], X[mask, 1], s=40, alpha=0.8, label=f'Class {j}')
ax.set_title('make_blobs Dataset — Ground Truth Labels', fontsize=13)
ax.set_xlabel('x1'); ax.set_ylabel('x2')
ax.legend()
plt.tight_layout()
plt.savefig('results/task2_dataset.png', dpi=150, bbox_inches='tight')
plt.show()
print("Plot saved.")
```

**Visual Verification:** The plot above visually confirms the underlying cluster geometry. This satisfies the isotropic cluster assumption critical for the squared Euclidean distance objective to yield optimal Voronoi partitions (Section 4, Eq. 9).
