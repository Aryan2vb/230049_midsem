# Task 1.3 — What the Paper Claims to Improve
**Paper:** Revisiting k-means: New Algorithms via Bayesian Nonparametrics (Kulis & Jordan, ICML 2012)

## Main Baseline

The primary baseline the paper compares against is **standard k-means (Lloyd's algorithm)**. The paper explicitly positions DP-means as a nonparametric extension of k-means and uses k-means as the direct algorithmic and empirical point of comparison throughout Section 3 and the experiments in Section 5.

---

## Limitation of the Baseline

Standard k-means requires the number of clusters **k to be fixed before the algorithm runs**. There is no principled, built-in mechanism in k-means to determine an appropriate value of k from the data itself. Practitioners must rely on external heuristics such as the elbow method, silhouette scores, or cross-validation — none of which are part of the algorithm and all of which require multiple runs with different k values.

---

## How DP-means Overcomes It

DP-means replaces the fixed-k requirement with a single penalty parameter **λ**, which implicitly controls the number of clusters. Whenever a point's distance to all existing cluster centers exceeds λ, a new cluster is created automatically. The value of k thus **emerges from the data** during a single run rather than being imposed externally. λ has a direct Bayesian interpretation as the penalty for creating a new cluster under a Dirichlet Process prior (Section 3), making it more principled than ad hoc k-selection heuristics.

---

## Condition Under Which DP-means Would NOT Outperform k-means

DP-means would not outperform standard k-means when the **true number of clusters k is already known with certainty**, the clusters are **spherical and well-separated** (satisfying the model assumptions of both methods equally), and the dataset is **large** (making per-point new-cluster checks expensive).

In this scenario, k-means with the correct k avoids the overhead of checking every point against the λ threshold and managing a dynamically growing center list. Both methods converge to the same local minimum of the same within-cluster sum of squares objective, but k-means does so with lower per-iteration cost and without the need to tune λ. The paper's advantage — automatic k discovery — provides zero benefit when k is already known, turning DP-means' added complexity into pure overhead.

This is not discussed explicitly by the authors, but follows directly from comparing Algorithm 1 of DP-means to Lloyd's algorithm: when λ = ∞ (or λ is set large enough that no new clusters are ever created), DP-means degenerates exactly to k-means initialized at x₁.
