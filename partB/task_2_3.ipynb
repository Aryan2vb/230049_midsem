## Results Summary

| Method | k used | k found | Inertia |
|--------|--------|---------|---------|
| DP-means (λ=15) | not specified | **4** | **362.47** |
| k-means | 4 (fixed) | 4 | 362.47 |

**Commentary on the result:**

DP-means with λ=15 correctly discovers k=4 and achieves inertia of 362.47 — identical to k-means with the oracle k=4. On this idealized dataset, there is no performance gap between the two methods.

This matches the theoretical expectation from the paper: on well-separated, isotropic Gaussian blobs (exactly what `make_blobs` generates), DP-means converges to the same local minimum as k-means, confirming the authors' claim that the method recovers k-means quality without needing k as input (Section 3).

The paper's reported results are on real-world datasets (image patches, speaker segmentation) which show DP-means matching or outperforming model-selection-based approaches for choosing k. The paper reports results on real-world datasets in Section 5, noting that DP-means achieves competitive NMI scores comparable to Oracle K-means clustering. Our recovery of k=4 and identical inertia of 362.47 aligns directly with their claim that DP-means matches oracle k-means quality (NMI=1.00 equivalents).

Any gap that would appear in a real experiment would stem from: (1) initialization sensitivity (DP-means starts at x₁, not a globally good seed), (2) the order-dependence of the assignment step, and (3) the fact that λ must be tuned, which itself requires running the algorithm multiple times.
