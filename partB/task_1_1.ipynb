# Task 1.1 — Core Contribution of Kulis & Jordan (2012)
**Paper:** Revisiting k-means: New Algorithms via Bayesian Nonparametrics

## Step-by-Step Description of DP-means (Algorithm 1)

---

### Step 1: Initialization
**What happens:** The algorithm begins with a single cluster whose center is set to the first data point x₁. The cluster count k is initialized to 1.

**Paper reference:** Algorithm 1, line 1 (Section 3)

**Purpose:** Unlike k-means, no value of k is required up front. The number of clusters grows dynamically as the data demands it.

---

### Step 2: Distance Computation
**What happens:** For each data point xᵢ in the dataset, the squared Euclidean distance is computed to every existing cluster center cⱼ.

**Paper reference:** Algorithm 1, line 4; Equation 9 (the DP-means objective function)

**Purpose:** This distance determines whether xᵢ belongs to any known cluster or warrants a new one. Squared Euclidean distance is used because DP-means is derived from a Gaussian mixture where all components share a spherical covariance (Section 4).

---

### Step 3: New Cluster Creation
**What happens:** If the minimum squared distance from xᵢ to all current centers exceeds the threshold λ, a new cluster is created with center initialized at xᵢ, and k is incremented by 1.

**Paper reference:** Algorithm 1, lines 5–6; λ is defined in Section 3 as the penalty for creating a new cluster, analogous to the concentration parameter α in the Dirichlet Process prior.

**Purpose:** λ is the single free parameter that controls the granularity of clustering. Larger λ = fewer, coarser clusters; smaller λ = more, finer clusters. This is the core innovation: k is not fixed but emerges from data and λ.

---

### Step 4: Assignment
**What happens:** If the minimum distance ≤ λ, the point xᵢ is assigned to the cluster with the nearest center (hard assignment).

**Paper reference:** Algorithm 1, lines 7–8

**Purpose:** Hard assignment is the small-variance (σ²→0) limit of the soft posterior assignments in a Dirichlet Process Gaussian Mixture Model (Section 4, Theorem 1). It produces a Voronoi partition of the space.

---

### Step 5: Center Update (M-step)
**What happens:** After all points are assigned, each cluster center is recomputed as the mean of all points currently assigned to it.

**Paper reference:** Algorithm 1, line 10

**Purpose:** This is the M-step analogue from EM — it minimizes the within-cluster sum of squared distances for the current assignment, driving the DP-means objective (Eq. 9) monotonically downward.

---

### Step 6: Convergence Check
**What happens:** Steps 2–5 repeat until the assignment of every point remains unchanged between consecutive iterations.

**Paper reference:** Section 3, convergence argument (the objective is bounded below and strictly decreasing each iteration, guaranteeing termination)

**Purpose:** The algorithm provably converges to a local minimum of the DP-means objective — a penalized variant of the k-means objective that includes a cost of kλ for using k clusters.

---

## Final Summary

This paper solves the problem of clustering without pre-specifying k by deriving DP-means from the small-variance asymptotic limit (σ²→0) of a Dirichlet Process Gaussian Mixture Model (DP-GMM), replacing the intractable nonparametric Bayesian model with a simple, deterministic, k-means-style algorithm where the number of clusters k is an emergent property of the data and the interpretable penalty parameter λ — which the authors claim makes their approach both more principled than heuristic k-selection methods and far more computationally efficient than full DP-GMM inference.
